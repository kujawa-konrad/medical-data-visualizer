Importing libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```



Importing data

```python
df = pd.read_csv('medical_examination.csv')
```

```python
df.head()
```

Adding 'overweight' column which inputs depend on calculated BMI

```python
df['weight']/(np.square(df['height']/100))
```

```python
df['overweight'] = np.where(df['weight']/(np.square(df['height']/100)) > 25, 1, 0)
```

```python
df.head()
```

Normalizing data by making 0 always good and 1 always bad. "If the value of 'cholesterol' or 'gluc' is 1, make the value 0. If the value is more than 1, make the value 1."

```python
df['cholesterol'] = np.where(df['cholesterol'] > 1, 1, 0)
```

```python
df.head()
```

```python
df['gluc'] = np.where(df['gluc'] > 1, 1, 0)
```

```python
df.head()
```

Creating new DataFrame in order to create categorical plot with one chart for each 'cardio' value.

```python
df_cat = df.melt(id_vars='cardio', value_vars = ['active', 'alco', 'cholesterol', 'gluc', 'overweight', 'smoke'])
```

```python
df_cat
```

```python
fig = sns.catplot(data=df_cat, kind='count', x='variable', hue='value', col='cardio').set(ylabel="total")
```

Cleaning data and drawing heatmap with upper triangle of correlation matrix masked. 

```python
df_heat = df[(df['ap_lo'] <= df['ap_hi'])&(df['height'] >= df['height'].quantile(0.025))&(df['height'] <= df['height'].quantile(0.975))&(df['weight'] >= df['weight'].quantile(0.025))&(df['weight'] <= df['weight'].quantile(0.975))]
df_heat
```

```python
corr = df_heat.corr()
corr
```

```python
mask = np.triu(np.ones_like(corr, dtype=bool))
```

```python
fig, ax = plt.subplots(figsize=(11, 9))
sns.heatmap(corr, mask=mask, vmax=.3, center=0, annot=True, fmt='.1f',
            square=True, cbar_kws={"shrink": .5, "ticks":[-0.08, 0.00, 0.08, 0.16, 0.24]})
```
